**Exercise 4:**

Analyse and run the following code. Identify Q1, Q2, Q3 in the boxplot. Change the dataset and discuss how the boxplot changes. Add a third variable.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Example data set with equal lengths for both variables
data = {
    'variable1': [70, 36, 43, 69, 82, 48, 34, 62, 35, 15, 59, 139, 46, 37, 42, 30, 55, 56, 36, 82,
                   38, 89, 54, 25, 35, 24, 22, 9, 56, 19],
    'variable2': [34, 67, 23, 45, 78, 90, 12, 58, 91, 23, 67, 89, 45, 31, 29, 76, 12, 47, 38, 59,
                   60, 42, 85, 20, 50, 75, 29, 18, 56, 45]  # Adjusted to be the same length
}

# Convert the data into a DataFrame
dataset = pd.DataFrame(data)

# Calculate statistics
mean = dataset.mean()
median = dataset.median()
variance = dataset.var()
std_dev = dataset.std()
min_values = dataset.min()
max_values = dataset.max()
data_range = max_values - min_values
quantiles = dataset.quantile([0.25, 0.5, 0.75])
length = dataset.count()

# Display statistics
print("Mean:\n", mean)
print("\nMedian:\n", median)
print("\nVariance:\n", variance)
print("\nStandard Deviation:\n", std_dev)
print("\nMinimum:\n", min_values)
print("\nMaximum:\n", max_values)
print("\nRange:\n", data_range)
print("\nQuantiles:\n", quantiles)
print("\nSample Size:\n", length)

# Create a summary DataFrame for the five-point summary (including mean)
summary = pd.DataFrame({
    'Min': min_values,
    '25th Percentile': quantiles.loc[0.25],
    'Median': quantiles.loc[0.5],
    '75th Percentile': quantiles.loc[0.75],
    'Max': max_values,
    'Mean': mean
})

print("\nFive-point summary for boxplot:\n", summary)

# Create a boxplot
plt.figure(figsize=(10, 5))
plt.boxplot([dataset['variable1'], dataset['variable2']], labels=['Variable 1', 'Variable 2'])
plt.title('Boxplot of Variables')
plt.ylabel('Values')
plt.grid()
plt.show()
```

**Exercise 5:**

Analyse the code below. Change the data sets and discuss how the correlation changes.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate example data
np.random.seed(42)
data = {
    'X': np.random.normal(0, 1, 100),  # Normal distribution (mean, standard deviation, the number of samples)
    'Y': np.random.normal(1, 2, 100),  # Normal distribution with different parameters
    'Z': np.random.uniform(-2, 2, 100)  # Uniform distribution (bound1, bound2, the number of samples)
}

# Convert data into a DataFrame
dataset = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = dataset.corr()

# Extract the correlation between X and Y
correlation_X_Y = correlation_matrix.loc['X', 'Y']
print(f"Correlation between X and Y: {correlation_X_Y:.4f}")

# Extract the correlation between Y and Z
correlation_Y_Z = correlation_matrix.loc['Y', 'Z']
print(f"Correlation between Y and Z: {correlation_Y_Z:.4f}")

# 1. Histogram
plt.figure(figsize=(10, 5))
plt.hist(dataset['X'], color='blue', bins=20, alpha=0.7, edgecolor='black')
plt.title('Histogram of X')
plt.xlabel('X values')
plt.ylabel('Frequency')
plt.grid()
plt.show()

# 2. Scatter plot for Y vs X with custom color and line width
plt.figure(figsize=(10, 5))
plt.scatter(dataset['X'], dataset['Y'], color='blue', s=50)  # s is the size of points
plt.title('Scatter Plot of Y vs X')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()

# 3. Adding another scatter plot for Z vs X
plt.figure(figsize=(10, 5))
plt.scatter(dataset['X'], dataset['Y'], color='blue', s=50, label='Y vs X')  # Scatter plot for Y vs X
plt.scatter(dataset['X'], dataset['Z'], color='green', s=50, label='Z vs X')  # Scatter plot for Z vs X
plt.title('Scatter Plot of Y and Z vs X')
plt.xlabel('X')
plt.ylabel('Y and Z')
plt.legend()
plt.grid()
plt.show()

# 4. Scatter plot matrix
sns.pairplot(dataset)
plt.suptitle('Scatter Plot Matrix', y=1.02)  # Adjust title position
plt.show()

```
